```python
import seaborn as sns

paleta = sns.color_palette('viridis')
```

```python
def filtro_por_palabra(columna, palabra):
    return df[columna].apply(lambda x: palabra in x)
```

```python
def graficar_filtros(filtro, etiquetas, titulo):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.pie(filtro, labels=etiquetas, autopct='%1.1f%%', colors=paleta)
    ax.set_title(titulo)
    plt.show()
```

```python
femenino = filtro_por_palabra('gender','female')
masculino = filtro_por_palabra('gender','male')
genero = [femenino.sum(), masculino.sum()]
labels_genero = ['Mujeres','Hombres']

graficar_filtros(genero, labels_genero,'GÃ©nero de Clientes\n perdidos')
```

```python
#Asistencia de IA
a_dos_anios = df[df['Contract'].apply(lambda x:'two' in x)]['tenure'].value_counts().sort_index()
a_un_anio = df[df['Contract'].apply(lambda x:'one' in x)]['tenure'].value_counts().sort_index()
mes_a_mes = df[df['Contract'].apply(lambda x:'month' in x)]['tenure'].value_counts().sort_index()
contrato = [a_dos_anios.sum(), a_un_anio.sum(), mes_a_mes.sum()]
labels_contrato = ['A dos aÃ±os','A un aÃ±o','Mes a mes']

fig, ax = plt.subplots(2, 2, figsize=(12, 8))

ax[0,0].pie(contrato, labels=labels_contrato, autopct='%1.1f%%', colors=paleta)
ax[0,0].set_title('Tipos de contrato')

ax[0,1].bar(a_dos_anios.index, a_dos_anios.values, color=paleta)
ax[0,1].set_title('AntigÃ¼edad en Contratos "Dos aÃ±os"')
ax[0,1].set_xlabel('Meses')
ax[0,1].set_ylabel('NÃºmero de clientes')

ax[1,0].bar(a_un_anio.index, a_un_anio.values, color=paleta)
ax[1,0].set_title('AntigÃ¼edad en Contratos Anuales')
ax[1,0].set_xlabel('Meses')
ax[1,0].set_ylabel('NÃºmero de clientes')

ax[1,1].bar(mes_a_mes.index, mes_a_mes.values, color=paleta)
ax[1,1].set_title('AntigÃ¼edad en Contratos "Mes a Mes"')
ax[1,1].set_xlabel('Meses')
ax[1,1].set_ylabel('NÃºmero de clientes')

plt.tight_layout()
plt.show()
```

```python
#Asistencia de IA
telefonÃ­a_consumida = df[df['PhoneService'].apply(lambda x: 'yes' in x if isinstance(x, list) else False)]['MultipleLines']
telefonÃ­a_no_consumida = df[df['PhoneService'].apply(lambda x: 'no' in x if isinstance(x, list) else False)]['MultipleLines']

serTelefonia = pd.concat([telefonÃ­a_consumida, telefonÃ­a_no_consumida]).value_counts()

serTelefonia_counts = serTelefonia.values
serTelefonia_labels = serTelefonia.index

traducciones = {
    ('yes',): 'Servicios de\n telefonÃ­a y \n multi-lÃ­neas',
    ('no',): 'Servicio de\n telefonÃ­a sin\n servicio de\n multi-lÃ­neas',
    ('no', 'phone', 'service'): 'Sin servicio\n de telefonÃ­a',
}

labels_traducidos = [traducciones.get(tuple(label), label) if isinstance(label, list) else traducciones.get(label, label) for label in serTelefonia_labels]

graficar_filtros(serTelefonia_counts, labels_traducidos, 'Servicio de telefonÃ­a\n y multi lÃ­neas')
```

```python
#Asistencia de IA
df_fiber = df[df['InternetService'].apply(lambda x: 'fiber' in x if isinstance(x, list) else False)]

fiber_security_yes = df_fiber['OnlineSecurity'].apply(lambda x: 'yes' in x if isinstance(x, list) else False).sum()
fiber_backup_yes = df_fiber['OnlineBackup'].apply(lambda x: 'yes' in x if isinstance(x, list) else False).sum()
fiber_protection_yes = df_fiber['DeviceProtection'].apply(lambda x: 'yes' in x if isinstance(x, list) else False).sum()

df_dsl = df[df['InternetService'].apply(lambda x: 'dsl' in x if isinstance(x, list) else False)]

dsl_security_yes = df_dsl['OnlineSecurity'].apply(lambda x: 'yes' in x if isinstance(x, list) else False).sum()
dsl_backup_yes = df_dsl['OnlineBackup'].apply(lambda x: 'yes' in x if isinstance(x, list) else False).sum()
dsl_protection_yes = df_dsl['DeviceProtection'].apply(lambda x: 'yes' in x if isinstance(x, list) else False).sum()

serInternet = [fiber_security_yes + dsl_security_yes,
               fiber_backup_yes + dsl_backup_yes,
               fiber_protection_yes + dsl_protection_yes]

serInternet_labels = ['Seguridad\n en lÃ­nea', 'Backup', 'Protection\n para\n dispositivos']

fig, ax = plt.subplots(figsize=(10, 6))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)



ax.barh(serInternet_labels, serInternet, color=paleta)
ax.set_title('Servicios adicionales')

for i,j in enumerate(serInternet):
    ax.text(j+20,i,str(j),color='black',fontsize=8.5,ha='left',va='center')

plt.show()
```
